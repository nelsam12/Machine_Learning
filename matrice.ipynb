```python
def multiplication(ligne:list, colonne:list):
    m = 0
    for i,j in zip(ligne, colonne):
        m += i*j
    return m

print(multiplication([1,2,3], [0,2,2]))
```

```python

class Matrice:
    def __init__(self:Matrice, valeur):
        self.lignes = len(valeur)
        self.colonnes = len(valeur[0])
        self.valeur = valeur

    def __str__(self:Matrice):
        m = "["
        for index,ligne in enumerate(self.valeur):
            m += str(ligne) if index == 0 else " "+ str(ligne)
            if index != self.lignes-1:
                m += "\n"
        m += "]"
        return m

    def transpose(self:Matrice):
        t_m = []
        for pc in range(self.colonnes):
            l_m = []
            for pl in range(self.lignes):
                l_m.append(self.valeur[pl][pc])
            t_m.append(l_m)
        return Matrice(t_m)

    def __add__(self1:Matrice, self2:Matrice):
        s_m = []
        for pl in range(self1.lignes):
            s_l = []
            for pc in range(self1.lignes): 
                s_l.append(self1.valeur[pl][pc] + self2.valeur[pl][pc])
            s_m.append(s_l)
        return Matrice(s_m)

    def __mul__(self1:Matrice, self2:Matrice):
        if self1.lignes != self2.colonnes: return "Opération impossible"
        m_m = []
        for ligne in self1.valeur:
            x = []
            for colonne in self2.transpose().valeur:
               x.append(multiplication(ligne, colonne))
            m_m.append(x)
        return Matrice(m_m)

    def __pow__(self:Matrice, n:float|int):
        if n == 0 : return self.__indentite() 
        if n == 1 : return self 
        return self * (self ** (n-1))


    def __indentite(self:Matrice):
        m_i = []
        for i in range(self.lignes):
            l = []
            for j in range(self.colonnes):
                l.append(1) if i == j else l.append(0)
            m_i.append(l)
        return Matrice(m_i)
                
           

m1 = Matrice([[0,2,-1],[0,0,1], [0,0,0]])
m2 = m1
print("Affichage de la matrice")
print(m1)
print("\n")
t_m1 = m1.transpose()
print("Affichage de la transposé")
print(t_m1)
print("\n")
print("Affiche de la somme")
print(m1 + m2)
print("\n")
print("Affiche du produit")
print(m1 * m2)
print("\n")
print("Affiche de la puissance")
print(m1 ** 1)

```
