```python
M = np.random.rand(3,4) * 10
print(M)
```

```python
def rotation(M):
    return np.rot90(np.rot90(M))
M_rotated = rotation(M)
print(M_rotated)
```

```python
def multiply(M1, M2):
    s = 0
    for i in range(len(M1)):
        L1 = M1[i]
        L2 = M2[i]
        ss = 0
        for j in range(len(L1)):
            ss += L1[j] * L2[j]
        s += ss
    return s
```

```python
def augmenter(M):
    x = M.shape
    M1 = np.zeros((x[0]+2,x[1]+2))
    M1[:, 1:-1][1:-1]=M
    return M1

print(augmenter(np.ones((2,2))))
```
